```python
import re
import string
import nltk    
from nltk.stem import WordNetLemmatizer
from nltk import tokenize    

f = open('lerolero/lerolero.txt', 'r')
text = f.read()
f.close()
lemmatizer = WordNetLemmatizer()
words = tokenize.word_tokenize(text.decode('utf-8'), language='portuguese')   
words = [lemmatizer.lemmatize(w).lower() for w in words if w not in string.punctuation]

frequency = {}
for word in words:
    count = frequency.get(word,0)
    frequency[word] = count + 1
frequency_words = sorted(frequency.items(), key=lambda x: x[1], reverse=True) # sorted by key, return a list of tuples
print(len(frequency_words))
```

```python
#!/usr/bin/env python
# -*- coding: utf-8 -*-
import matplotlib.pylab as plt
my_xticks, y = zip(*frequency_words) # unpack a list of pairs into two tuples
my_xticks = [t for t in my_xticks]
x = range(len(my_xticks))
plt.xticks(x[:20], my_xticks[:20],rotation=45)
plt.plot(x[:20], y[:20])
plt.show()
```
