# Prepare stop words process

```python
import nltk
nltk.download('stopwords')
nltk.download('punkt')
```

```python
from nltk.corpus import stopwords
```

```python
from string import punctuation
```

```python
from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize
```

```python
stopwords = set(stopwords.words("portuguese") + list(punctuation))
```

```python
def remove_stopwords(text):
  palavras = word_tokenize(text.lower(), language='portuguese')
  palavras_sem_stopwords = [palavra for palavra in palavras if palavra not in stopwords]
  return " ".join(palavras_sem_stopwords)
```

#Get corpus from sentences of a book

```python
sent_tokenizer = nltk.data.load('tokenizers/punkt/portuguese.pickle')
corpus = sent_tokenizer.tokenize(book)
corpus
```

```python
len(corpus)
```

#Prepare Lemmatization process

### There is not nltk lematization for portuguese indeed there are others libraries which can lemmatize in portuguese language.
https://lars76.github.io/2018/05/08/portuguese-lemmatizers.html

```python
!pip install git+https://github.com/stanfordnlp/stanza.git
```

```python
import stanza
stanza.download('pt')
nlp = stanza.Pipeline('pt')
```

```python
def lemmatizacao(corpus):
  text = ""
  pos = ""
  lemma = ""
  for sent in nlp(corpus).sentences:
    for word in sent.words:
        text += word.text + " "
        pos += word.upos + " "
        lemma += word.lemma + " "
  return text, pos, lemma
```

# Processing corpus

```python
sem_stopwords_corpus = []
for text in corpus:
  text = remove_stopwords(text)
  if text != "":
    sem_stopwords_corpus.append(text) 
sem_stopwords_corpus
```

```python
len(sem_stopwords_corpus)
```

```python
lemma_corpus = []
for text in sem_stopwords_corpus:
  text,pos,lemma = lemmatizacao(text)
  if text != "":
    lemma_corpus.append(lemma) 
lemma_corpus
```

# TFIDF process

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(min_df=1)
X = vectorizer.fit_transform(lemma_corpus)
idf = vectorizer.idf_
print(dict(zip(vectorizer.get_feature_names_out(), idf))) #show idf of each word
```
